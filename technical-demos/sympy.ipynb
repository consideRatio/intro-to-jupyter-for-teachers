# Sympy
## About
Sympy is a _Computer Algebra System_ (CAS) that help you do algebra. Sympy can help you if you want to...
- solve an equation algebraically
- differentiate a function algebraically
- simplify an expression
- find the lowest common denominator
- plot expressions

#### Keywords
- [CAS](https://en.wikipedia.org/wiki/Computer_algebra_system)

## Example

```python
import sympy as sp # !pip install sympy
sp.init_printing() # enables latex printout of expressions

x, y, z = sp.symbols('x, y, z') # get variables containing symbols that sympy understands
```

```python
# create an expression as a combination of symbols
exp = (x/3)**2 - 1
exp
```

```python
sp.plot(exp);
```

```python
# get the derivative
derivated_exp = sp.diff(exp)
derivated_exp
```

```python
sp.plot(derivated_exp);
```
