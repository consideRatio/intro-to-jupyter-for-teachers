# Jupyter
## About
Jupyter notebooks are `.ipynb` files that are viewed through a web browser hosted by a _notebook server_, allowing the users to run various types of code depending on what kind of _kernel_ is used. The notebooks contains both _markdown cells_ and _code cells_. The code cells can when run give you various outputs, such as text, images or HTML.

#### Keywords
- notebook
- markdown cell
- code cell
- output
- notebook server
- kernel
